```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the Excel file
data = pd.read_excel('pokemon_data.xlsx')
```

```python
data.columns
```

# ● Get all Pokemons whose spawn rate is less than 5%

```python
# Filter the data for Pokémon with spawn rate less than 5%
low_spawn_rate_pokemon = data[data['spawn_chance'] < 5]

# Display the filtered data
print("Pokémon with spawn rate less than 5%:")

# Plot the spawn rate distribution
plt.figure(figsize=(10, 6))
plt.hist(data['spawn_chance'], bins=10, edgecolor='black')
plt.axvline(x=5, color='red', linestyle='--', label='5%')
plt.title('Spawn Rate Distribution')
plt.xlabel('Spawn Rate')
plt.ylabel('Count')
plt.legend()
plt.show()
```

# Get all Pokemons that have less than 4 weaknesses

```python
# Filter the data for Pokémon with less than 4 weaknesses
pokemon_less_than_4_weaknesses = data[data['weakness'].apply(lambda x: len(x) < 4)]

# Display the filtered data
print("Pokémon with less than 4 weaknesses:")
print(pokemon_less_than_4_weaknesses)

# Plot the count of Pokémon with different numbers of weaknesses
plt.figure(figsize=(10, 6))
pokemon_weaknesses_count = data['weakness'].apply(lambda x: len(x))
plt.hist(pokemon_weaknesses_count, bins=range(10), edgecolor='black')
plt.xlabel('Number of Weaknesses')
plt.ylabel('Count')
plt.title('Pokémon Weaknesses Distribution')
plt.show()
```

#  Get all Pokemons that have no multipliers at all 

```python
# Plot the count of Pokémon with and without multipliers
plt.figure(figsize=(10, 6))
pokemon_multipliers_count = data['multipliers'].apply(lambda x: len(x))
pokemon_no_multipliers_count = len(data) - pokemon_multipliers_count.sum()
plt.bar(['With Multipliers', 'Without Multipliers'], [pokemon_multipliers_count.sum(), pokemon_no_multipliers_count])
plt.xlabel('Pokémon')
plt.ylabel('Count')
plt.title('Pokémon with and without Multipliers')
plt.show()
```

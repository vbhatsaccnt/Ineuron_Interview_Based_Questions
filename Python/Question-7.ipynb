```python
import pandas as pd

# Load data from Excel file
data = pd.read_csv("meteorite_data.csv")

# Display the loaded data
print(data.head())
```

```python
# Split the year column and create a new column
data[['year', 'month', 'day']] = data['year'].str.split('-', expand=True)

# Display the updated data
print(data.head())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Convert the 'year' column to datetime format
#data['year'] = pd.to_datetime(data['year'], errors='coerce')

# Filter the data for Earth meteorites before the year 2000
filtered_data = data[(data['nametype'] == 'Valid') & (data['year'] < '2000-01-01')]

# Create a count plot of meteorite counts by year
plt.figure(figsize=(10, 6))
sns.countplot(data=filtered_data, x='year')
plt.xlabel('year')
plt.ylabel('Meteorite Count')
plt.title('Frequency of Earth Meteorites before 2000')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter the data for Earth meteorites with a mass greater than 10000 kg
filtered_data = data[(data['nametype'] == 'Valid') & (data['year'] < '2000-01-01') & (data['mass'] > 10000)]

# Create a scatter plot of mass vs. year for the filtered data
plt.scatter(filtered_data['year'], filtered_data['mass'])
plt.xlabel('Year')
plt.ylabel('Mass (kg)')
plt.title('Earth Meteorites with Mass > 10000 kg')
plt.show()
```
